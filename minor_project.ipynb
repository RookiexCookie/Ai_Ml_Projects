##  Hospital Appointment Cancellation Analysis
#####1. From: Minor 1 â€“ Pandas

#####2. Extension: Analyze cancellation patterns
#####3. Add-ons: Bar plots, summary stats
#####4. Difficulty: Easy



```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("hospital.csv")
df.head()
```

```python
cancellation_rate = df.cancelled.mean()
cancellation_rate
```

```python
waiting_cancellation = df.wait_days.corr(df.cancelled)
waiting_cancellation
```

```python
dept_cancel = df.groupby("department")["cancelled"].mean()
dept_cancel
```

```python
dept_cancel.plot(kind="bar")
plt.ylabel("Cancellation Rate")
plt.title("Cancellation Rate by Department")
plt.show()
```
